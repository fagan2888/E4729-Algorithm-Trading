```python
import warnings
warnings.filterwarnings('ignore')

import os
import time
import pickle
import numpy as np
import pandas as pd
from datetime import datetime

import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
%matplotlib inline
```

```python
path = os.getcwd()
```

```python
train_start_time = '2019-01-01 00:00:00'
train_end_time = '2019-09-10 21:59:59'
test_start_time = '2019-09-10 22:00:00'
test_end_time = '2020-01-01 00:00:00'
```

```python
prod = 'BTCUSD'
price = pd.read_csv(path + '/clean_data/hourly%s-H1.csv' % prod, sep='\t')
price['ds'] = pd.to_datetime(price['Date'].astype(str) + ' ' + price['Time'], format='%Y%m%d %H:%M:%S')
price.index = pd.DatetimeIndex(price['ds'])

data = pd.read_csv(path + '/result/%s-forecast_df.csv' % prod)
data['ds'] = pd.to_datetime(data['ds'], format='%Y-%m-%d %H:%M:%S')
data.index = pd.DatetimeIndex(data['ds'])
```

```python
price.head()
```

```python
data.head()
```

```python
data['pred_larger_than_zero'] = data['yhat'] > 0
data['pred_smaller_than_zero'] = data['yhat'] < 0
```

```python
periods = 3

data['strategy'] = 0
data['strategy'].loc[data['pred_larger_than_zero'].cumsum().diff(periods=periods).fillna(method='bfill') == periods] = 1
data['strategy'].loc[data['pred_smaller_than_zero'].cumsum().diff(periods=periods).fillna(method='bfill') == periods] = -1
```

```python
data['strategy']
```

```python
data = pd.concat([price, data], axis=1).dropna(axis=0)
```

```python
data
```

```python
useCols = ['Open', 'yhat', 'strategy']

train = data.loc[train_start_time: train_end_time, useCols]
test = data.loc[test_start_time: test_end_time, useCols]
```

```python
base = 10000
tmp = train
hold = 0
deposit = 10000
returnList = []
previousPrice = 0
numTrade = 0

for date, s in tmp.iterrows():
    p, yhat, strategy = s
    
    if strategy == 1:
        
        if hold in [0]:
            print('%s: 買進 - 現價 %.f' % (date, p))
            # buy one unit
            deposit -= p
            previousPrice = p
        hold = 1
        
        
    elif strategy == -1:
        if hold in [0]:
            print('%s: 買空 - 現價 %.2f' % (date, p))
            # sell one unit
            deposit += p
            previousPrice = p
        hold = -1
    
    else:
        if hold == 1:
            print('%s: 賣出 - 現價 %.2f - 損益 %.5f' % (date, p, p - previousPrice))
            # sell one unit
            returnList.append(p - previousPrice)
            deposit += p
            numTrade += 1

        elif hold == -1:
            print('%s: 回補 - 現價 %.2f - 損益 %.5f' % (date, p, previousPrice - p))
            # buy one unit
            returnList.append(previousPrice - p)
            deposit -= p
            numTrade += 1
        hold = 0
        
    
if hold == 1:
    # sell one unit
    returnList.append(p - previousPrice)
    deposit += p
elif hold == -1:
    # buy one unit
    returnList.append(previousPrice - p)
    deposit -= p
    
print('number of trade: ', numTrade)
```

```python
buyHold = round((train['Open'].iloc[-1] - train['Open'].iloc[0]), 3)
print('Total Return: ', round(sum(returnList), 3))
print('Return Rate: ', round((deposit - base)/base, 3))
print('Buy and Hold: ', buyHold)


f, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8), sharex=True, gridspec_kw={'height_ratios': [3, 1]})
ax1.plot(pd.Series(returnList).cumsum(), label='Strategy')
ax1.set_title('Cumulative Return in Training Set')
ax1.set_ylabel('dollar')
ax1.axhline(buyHold, color='r', label='Buy and Hold Return: %s' % buyHold)
ax1.axhline(0, color='black', linestyle='--', alpha=0.8)
ax1.legend()
ax2.bar(list(range(len(returnList))), returnList)
ax2.set_ylabel('dollar')
ax2.set_title('return in each transaction')
```
